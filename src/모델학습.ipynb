```python
# 아까 Roboflow에서 복사한 '다운로드 코드'를 아래에 그대로 붙여넣고 실행하세요!
# (예시 코드입니다. 학생들은 본인들의 코드로 바꿔야 합니다.)
!pip install roboflow
from roboflow import Roboflow
rf = Roboflow(api_key="여기에_복사한_키를_넣으세요")
project = rf.workspace("workspace-name").project("project-name")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
from ultralytics import YOLO

# 가장 작고 빠른 Nano 모델을 도화지로 사용합니다.
model = YOLO('yolov8n.pt') 

# AI 훈련 시작! (epochs=50 은 사진들을 50번 반복해서 공부하라는 뜻입니다.)
# dataset.location 경로 안의 data.yaml 파일을 교과서로 삼습니다.
results = model.train(data=dataset.location + '/data.yaml', epochs=50, imgsz=640)
```
